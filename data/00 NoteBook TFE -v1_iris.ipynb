## Preparación de dataSets

Se hace la importación del DataSet obtenido de Facebook, con técnica de Web Scraping.

```
## Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
```

```
data_facebook = pd.read_json(r'201209_json_result_scraping.json')
data_facebook.head()
```

Se convirten las variables que son categoricas en variables numéricas.

```
iris = load_iris()
```

```
# iris.keys()
# iris['data'][0]
target_names = iris['target_names']
iris['target']
```

```
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'])
```

```
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
```

```
knn.score(X_test, y_test)
```

```
prediction = knn.predict([[3.2,1.4,2.6,1.1]])
print('*'*70)
print(target_names[prediction])
```
